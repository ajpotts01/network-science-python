```python
# Okinawa news
url_jp_news: str = "http://english.ryukyushimpo.jp/"
data_jp_news: str = requests.get(url=url_jp_news).text

data_jp_news
```

```python
# Parse above HTML with bs4
soup: BeautifulSoup = BeautifulSoup(markup=data_jp_news, features="html.parser")
links: ResultSet[Any] = soup.find_all(name="a", href=True)
links
```

```python
# Better URL extraction
urls: list[str] = [link.get("href") for link in links]
urls
```

```python
# Get specific page
url_article: str = "http://english.ryukyushimpo.jp/2021/09/03/34020/"
data_article: str = requests.get(url=url_article).text

soup: BeautifulSoup = BeautifulSoup(markup=data_article, features="html.parser")
text_article: str = soup.get_text()
```

```python
# Cleaning
chr_space: str = " " 
text_cleaned: str = (
    text_article
    .replace("\n", chr_space)
    .replace("\r", chr_space)
    .replace("\t", chr_space)
    .replace("\xa0", chr_space)
)

index_start: int = text_cleaned.rindex("Hirokazu Ueyonabaru")
index_end: int = text_cleaned.index("Go to Japanese")
text_trimmed: str = text_cleaned[index_start:index_end]
text_trimmed
```
